```python
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf
from matplotlib import pyplot

autocorrelation = []


a = np.array([1,4,2,4,5,8,7,4,7,9,1,3,5,6])


for shift in range(10):
    l, r = a[:-shift if shift>0 else len(a)], a[shift:]
    print(l, r)
    correlation = np.corrcoef(l, r)[0, 1]
    autocorrelation.append(correlation)


print(autocorrelation)

plot_acf(a, lags=9)
pyplot.show()

# https://www.helenkapatsa.ru/avtokorrieliatsiia/
```
